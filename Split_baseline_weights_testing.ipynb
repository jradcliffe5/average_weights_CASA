```python
## Test 4 antennas with antenna weights
ant_weights = [1,20,40,50]

## Measured baseline weights are:
c01 = ant_weights[0]*ant_weights[1]
c02 = ant_weights[0]*ant_weights[2]
c03 = ant_weights[0]*ant_weights[3]
c13 = ant_weights[1]*ant_weights[3]
c12 = ant_weights[1]*ant_weights[2]
c23 = ant_weights[2]*ant_weights[3]

## Define functions where product of many variables = sum of baseline weights... 4 equations to solve!
def func_test(x):
    out = [x[0]*(x[1]+x[2]+x[3]) - c01 - c02 - c03] 
    out.append(x[1]*(x[0]+x[2]+x[3]) - c01 - c12 - c13)
    out.append(x[2]*(x[0]+x[1]+x[3]) - c02 - c12 - c23)
    out.append(x[3]*(x[0]+x[1]+x[2]) - c03 - c13 - c23)
    return out
from scipy.optimize import fsolve, lsq_linear
import numpy as np

## Put this into scipy solver
x0 = fsolve(func_test, np.ones(4))
```
